```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

As some methods have less than 100 discoveries in total, we will group all of them in a group names "Other", for this we will use a conditional while grouping data.

```python
d = pd.read_csv("../data/data.csv")

#print(d.columns)
methods = d.discoverymethod.unique()
method_groups = []
method_counts = []
other_group = []
m = []
for i in methods:
    df = d[d.discoverymethod == i]
    if len(df) < 70:
        other_group.append(df)
    else:
        m.append(i)
        method_groups.append(df)
        method_counts.append(len(df))
        
other = pd.concat(other_group, ignore_index = True, axis = 0)
method_groups.append(other)
m.append("Others")
method_counts.append(len(other))
met = pd.DataFrame(list(zip(m, method_counts)), columns = ["Method", "Number"])
met
```

## 1. All methods by year

```python
fig1, ax = plt.subplots()
ax.hist(d.disc_year)
ax.set_ylabel("Number of exoplanets")
ax.set_xlabel("Year of discovery")
fig1.savefig("../figs/hist_year.png", dpi = 300)
```

Jmmm, something is wrong, we are not getting one bar per year (the default value of hist bars is 10), so we need to calculate how many years have been passed since the first discovery, we can do this by getting the unique values from the column `disc_year`

```python
fig1, ax = plt.subplots()
bns = len(d.disc_year.unique())
yearmin = d.disc_year.min()
yearmax = d.disc_year.max()
bns = range(yearmin, yearmax)
ax.hist(d.disc_year, bins = bns)
ax.set_ylabel("Number of exoplanets")
ax.set_xlabel("Year of discovery")
fig1.savefig("../figs/hist_year.png", dpi = 300)
```

## 2. Histograms by method
Now we can answer the question, are these peaks related to a certain method? We can do a histogram for each method, superpose them or do single plots for each one.

```python
fig1, ax = plt.subplots()
for i, j in zip(m, method_groups):
    #bns = len(j.disc_year.unique())
    ax.hist(j.disc_year, bins = bns, label = i)
ax.set_ylabel("Number of exoplanets")
ax.set_xlabel("Year of discovery")
ax.legend()
fig1.savefig("../figs/hist_year_method.png", dpi = 300)
```

### Habitable exoplanets
For this category we have a special column names `hwc`, on which we have a value of 1 if the exoplanet is considered habitable or 0 if else. Then we will only filter this category and repeat the same plot

```python
fig1, ax = plt.subplots()
for i, j in zip(m, method_groups):
    #bns = len(j.disc_year.unique())
    hab = j[j.hwc == 1]
    ax.hist(hab.disc_year, bins = bns, label = i)
ax.set_ylabel("Number of exoplanets")
ax.set_xlabel("Year of discovery")
ax.legend()
#fig1.savefig("../figs/hist_year_method.png", dpi = 300)
```

Another way of seeing this is doing 4 separate plots that share the same x axis.

```python
fig1, ax = plt.subplots(4,1, sharex = True)
axc = 0
for i, j in zip(m, method_groups):
    #bns = len(j.disc_year.unique())
    ax[axc].hist(j.disc_year, bins = bns, label = i)
    ax[axc].legend()
    axc += 1
ax[-1].set_xlabel("Year of discovery")
fig1.savefig("../figs/hist_year_methodmtx.png", dpi = 300)
```

Now we will focus on how far from its host star is the exoplanet located, and we will do histograms of this parameter by method. On the dataframe the column of semi major axis in Astronomical Units (AU) is named `pl_orbsmax`.


```python
fig1, ax = plt.subplots(4,1, sharex = True)
axc = 0
xmax = j.pl_orbsmax.max()
xmin = j.pl_orbsmax.min()
bns = np.linspace(xmin, xmax, 15)
for i, j in zip(m, method_groups):
    ax[axc].hist(j.pl_orbsmax, bins = bns, label = i)
    ax[axc].legend()
    axc += 1
ax[-1].set_xlabel("Semi major axis [au]")
#ax[-1].set_xlim(0,xmax+2)
#fig1.savefig("../figs/hist_year_methodmtx.png", dpi = 300)
```

Why is our plot so long? There seems to be a planet or two measured at more than 7000 astronomical units, this won't be taken into account right now, we will focus on study a closer range of the star, for example 25AU. Then, all planets' orbits that have a semi major axis greater than this number will be taken into account as 25 AU. Also, don't forget to _drop_ the values that are not available (NaN), we use the method `dropna()` for this.

To get this, we will use a mask, we can do this in two ways (can be read in [Pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.where.html))
 - Using `where`, we use a condition and if the cond is not met, we replace cet value with another
 - Using `mask`, the same as `where` but the value is replaced if the condition is met

```python
fig1, ax = plt.subplots(4,1, sharex = True)
axc = 0
xmax = 25 # range of study
bns = np.linspace(0, xmax) # by default creates bars of 1AU
for i, j in zip(m, method_groups):
    x = j.pl_orbsmax.dropna().where(j.pl_orbsmax < xmax, xmax)
    ax[axc].hist(x, bins = bns, label = i)
    ax[axc].legend()
    axc += 1
ax[-1].set_xlabel("Semi major axis [AU]")
ax[-1].set_xlim(0,xmax+2) #you can change the limit of the plot
fig1.savefig("../figs/hist_AU_methodmtx.png", dpi = 300)
```

## 3. Planet radius histograms
Now we will focus on how big is the exoplanet, and we will do histograms of this parameter by method. On the dataframe the column of the planet radius is `pl_rade` (in Earth radius) or `pl_radj` in Jupiter radius.


```python
fig1, ax = plt.subplots(4,1, sharex = True)
axc = 0
xmax = 20 # range of study
bns = np.linspace(0, xmax, 20) # by default creates bars of 1AU
print(bns)
for i, j in zip(m, method_groups):
    x = j.pl_rade.where(j.pl_rade < xmax, xmax)
    ax[axc].hist(x, bins = bns, label = i)
    ax[axc].legend()
    axc += 1
ax[-1].set_xlabel("Radius of the planet (Earth Radius)")
ax[-1].set_xlim(0,xmax+2) #you can change the limit of the plot
#fig1.savefig("../figs/hist_AU_methodmtx.png", dpi = 300)
```
